There are some NaN values in form of text and some unwanted string characters

```python
df['Time_taken(min)'] = df['Time_taken(min)'].str.replace('(min) ', '', regex=False)
df['Time_taken(min)']
```

regex=False treats the pattern (the part of string we want to replace) as a literal string

Converting NaN strings into actual null values

```python
for i in df.columns:
    df[i].loc[(df[i] == 'NaN') | (df[i] == 'NaN ') | (df[i] == 'conditions NaN')]=np.nan

df.isnull().sum()
```

```python
for i in dftest.columns:
    dftest[i].loc[(dftest[i] == 'NaN') | (dftest[i] == 'NaN ') | (dftest[i] == 'conditions NaN')]=np.nan
```

Converting some features to numeric format

```python
df['Delivery_person_Age']=pd.to_numeric(df['Delivery_person_Age'])
df['Delivery_person_Ratings']=pd.to_numeric(df['Delivery_person_Ratings'])
df['Time_taken(min)']=pd.to_numeric(df['Time_taken(min)'])

df.info()
```

```python
dftest['Delivery_person_Age']=pd.to_numeric(dftest['Delivery_person_Age'])
dftest['Delivery_person_Ratings']=pd.to_numeric(dftest['Delivery_person_Ratings'])
```

Dealing with missing data

```python
df.dropna(subset=['Time_Orderd'],inplace=True)
df['Delivery_person_Age'].fillna(df['Delivery_person_Age'].mean(), inplace=True)
df['Delivery_person_Ratings'].fillna(df['Delivery_person_Ratings'].mean(), inplace=True)
df['Weatherconditions'].fillna(df['Weatherconditions'].mode(), inplace=True)
df['Road_traffic_density'].fillna(df['Road_traffic_density'].mode(), inplace=True)
df['multiple_deliveries'].fillna(df['multiple_deliveries'].mode(), inplace=True)
df['Festival'].fillna(df['Festival'].mode(), inplace=True)
df['City'].fillna(df['City'].mode(), inplace=True)
df.isnull().sum()
```

```python
dftest.dropna(subset=['Time_Orderd'],inplace=True)
dftest['Delivery_person_Age'].fillna(dftest['Delivery_person_Age'].mean(), inplace=True)
dftest['Delivery_person_Ratings'].fillna(dftest['Delivery_person_Ratings'].mean(), inplace=True)
dftest['Weatherconditions'].fillna(dftest['Weatherconditions'].mode(), inplace=True)
dftest['Road_traffic_density'].fillna(dftest['Road_traffic_density'].mode(), inplace=True)
dftest['multiple_deliveries'].fillna(dftest['multiple_deliveries'].mode(), inplace=True)
dftest['Festival'].fillna(dftest['Festival'].mode(), inplace=True)
dftest['City'].fillna(dftest['City'].mode(), inplace=True)
dftest.isnull().sum()
```

## Model Training

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from xgboost.sklearn import XGBRegressor
from sklearn import metrics
#from sklearn.model_selection import GridSearchCV
```

```python
x = df.drop('Time_taken(min)', axis=1)
y = df['Time_taken(min)']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```

### Random Forest

Training

```python
rf=RandomForestRegressor(n_estimators=100, max_depth=8)
```

```python
rf.fit(x_train,y_train)
y_pred_rf = rf.predict(x_test)
y_pred_rf
```

Evaluation

```python
print('RMSE of RF:', ( metrics.mean_squared_error(y_test,y_pred_rf) )**0.5 )
print('R2 of RF:', metrics.r2_score(y_test,y_pred_rf) )
```

### XGBoost

Training

```python
xg=XGBRegressor(n_estimators=100, max_depth=8)
```

```python
xg.fit(x_train,y_train)
y_pred_xg = xg.predict(x_test)
y_pred_xg
```

Evaluation

```python
print('RMSE of XGBoost:', ( metrics.mean_squared_error(y_test,y_pred_xg) )**0.5 )
print('R2 of XGBoost:', metrics.r2_score(y_test,y_pred_xg) )
```

### Model Deployment
